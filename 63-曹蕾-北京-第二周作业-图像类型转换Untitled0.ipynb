```python

from skimage.color import rgb2gray
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2


#灰度化
img = cv2.imread("/content/drive/MyDrive/Colab_Notebooks/Code/input/test_images/10078.tiff")
h,w = img.shape[:2]
img_gray = np.zeros([h,w],img.dtype)
img_gray1 = np.zeros([h,w],img.dtype)#二值化空矩阵
for i in range(h):
  for j in range(w):
    m = img[i,j]
    img_gray[i,j] = int(m[0]*0.11 + m[1]*0.59 + m[2]*0.3)
# from google.colab.patches import cv2_imshow
# print(img_gray)


# from google.colab.patches import cv2_imshow
plt.subplot(221)
plt.imshow(img_gray,cmap='gray')
plt.subplot(222)
plt.imshow(img)
print("--image 10078--")
print(img)
print("image show gray: %s"%img_gray)
img_gray1 = np.float32(img_gray/255)

# cv2.imshow("image show gray",img_gray)
#二值化：
rows, cols = img_gray.shape
# for i in range(rows):
#   for j in range(cols):
#     if(img_gray[i,j] <= 0.5):
#       img_gray[i,j] = 0
#     else:
#       img_gray[i,j] = 1
for i in range(h):
  for j in range(w):
    img_gray1[i,j] = np.float32(img_gray[i,j]/255)

# print("img_gray1:/255")  
print(img_gray1)   
img_binary = np.where(img_gray1 >= 0.5, 1, 0)

for i in range(h):
  for j in range(w):
    img_binary[i,j] = int(img_binary[i,j]*255)

print("--image binary--")
print(img_binary)
print(img_binary.shape)

plt.subplot(223)
plt.imshow(img_binary, cmap='gray')
plt.show()
```
